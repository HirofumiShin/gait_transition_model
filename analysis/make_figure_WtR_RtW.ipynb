## Import module

````python
import numpy as np
import matplotlib.pyplot as plt
import sys, os
sys.path.append(os.path.join(os.getcwd(), '../'))
from src.config import *
import pandas as pd
from matplotlib.ticker import MaxNLocator


plt.rcParams['font.family'] = 'Times New Roman'
data_folder = str(os.getcwd())+"/../data"
````

## Set label

````python
phase_mapping_actual = {
    "PHASE.JUMPING": 2,          # ジャンピングは上の方
    "PHASE.SINGLE_SUPPORT": 1,   # シングルサポートは中間
    "PHASE.DOUBLE_SUPPORT": 0    # ダブルサポートを一番下に
}

phase_mapping_virtual = {
    "PHASE.JUMPING": 4,          # ジャンピングは上の方
    "PHASE.SUPPORT": 3
}

phase_mapping = {
    "PHASE.JUMPING": 2,          # ジャンピングは上の方
    "PHASE.SINGLE_SUPPORT": 1,   # シングルサポートは中間
    "PHASE.DOUBLE_SUPPORT": 0,    # ダブルサポートを一番下に
    "PHASE.JUMPING": 4,          # ジャンピングは上の方
    "PHASE.SUPPORT": 3
}

rename_list_actual=['DS Phase', 'SS Phase', 'F Phase']

rename_list_virtual=['SS Phase', 'F Phase']


rename_list=['DS Phase', 'SS Phase', 'F Phase', r'SS$_{v}$ Phase',  r'F$_{v}$ Phase']
````

## Phase figure with gait speed from low to high

````python
data = pd.read_csv(str(data_folder) + "/gait_transition_from_low_to_high.csv")

data = data.T
start = 22102
end = start+600
time = data.iloc[0][start:end]
y_phase = [phase_mapping_actual.get(phase, -1) for phase in data.iloc[28][start:end]]

fig, ax = plt.subplots(figsize=(2, 1.5))
plt.plot(time, y_phase, label='Column 28')
y_phase = [phase_mapping_virtual.get(phase, -1) for phase in data.iloc[30][start:end]]
plt.plot(time, y_phase, label='Column 28')
ax.set_yticks([0, 1,2,3,4])
ax.set_yticklabels(rename_list)
plt.xlabel('Time (s)')
plt.ylabel('Phase')
ax.xaxis.set_major_locator(MaxNLocator(5))
plt.grid()
plt.savefig("../fig/test1.pdf")
plt.show()


###

start = 119885+62
end = start+140
time = data.iloc[0][start:end]
y_phase = [phase_mapping_actual.get(phase, -1) for phase in data.iloc[28][start:end]]


fig, ax = plt.subplots(figsize=(2, 1.5))
plt.plot(time, y_phase, label='Column 28')
####
y_phase = [phase_mapping_virtual.get(phase, -1) for phase in data.iloc[30][start:end]]
plt.plot(time, y_phase, label='Column 28')
ax.set_yticks([0, 1,2,3,4])
ax.set_yticklabels(rename_list)
plt.xlabel('Time (s)')
plt.ylabel('Phase')
ax.xaxis.set_major_locator(MaxNLocator(5))
plt.grid()
plt.savefig("../fig/test2.pdf")
plt.show()
````
